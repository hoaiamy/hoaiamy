```python
import pandas as pd

tx_2017=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2017_tx_all-records_labels.csv")
tx_2016=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2016_tx_all-records_labels.csv")
tx_2015=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2015_tx_all-records_labels.csv")
tx_2014=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2014_tx_all-records_labels.csv")
tx_2013=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2013_tx_all-records_labels.csv")
tx_2012=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2012_tx_all-records_labels.csv")
tx_2011=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2011_tx_all-records_labels.csv")
tx_2010=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2010_tx_all-records_labels.csv")
tx_2009=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2009_tx_all-records_labels.csv")
tx_2008=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2008_tx_all-records_labels.csv")
tx_2007=pd.read_csv(r"C:\Users\ngphu\Downloads\hmda_2007_tx_all-records_labels.csv")
```

```python
merge_cols = [
    "as_of_year",
    "respondent_id",
    "loan_type_name",
    "loan_purpose_name",
    "loan_amount_000s",
    "action_taken_name",
    "applicant_income_000s",
    "purchaser_type_name",
    "agency_name",
    "property_type_name",
    "lien_status_name",
    "hud_median_family_income",
]

sample_n = 1000  # change this to the maximum rows to take from each year

year_dfs = [
    tx_2007,
    tx_2008,
    tx_2009,
    tx_2010,
    tx_2011,
    tx_2012,
    tx_2013,
    tx_2014,
    tx_2015,
    tx_2016,
    tx_2017,
]

dfs = []
for df in year_dfs:
    n_rows = min(sample_n, len(df))
    dfs.append(df.head(n_rows)[merge_cols])

# Combine the chosen rows and keep unique key combinations.
tx_subset = pd.concat(dfs, ignore_index=True)
tx_merged = tx_subset.drop_duplicates(subset=merge_cols).reset_index(drop=True)

print(f"Selected rows per year capped at {sample_n}; merged shape: {tx_merged.shape}")
```
